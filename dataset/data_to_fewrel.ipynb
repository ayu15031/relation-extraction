```python
import json
import string

jsons = {"bio_data_predict_filtered_58.json": "noun/symptom/covid", "locationlocationcontains.json": "location/location/contains", "businesspcompany.json": "business/person/company", "peoplepeopletitle.json": "people/people/title", "peoplelocationlivesin.json": "people/location/lives_in", "peopleviruspositive.json": "people/virus/positive", "personpersonrelation.json": "person/person/relation"}

final_list = {}
for file in jsons:
    
    f = open(file, "r")

    final_list[jsons[file]] = []

    for a in f:
        try:
            obj = json.loads(a)


            if "_" in obj['head']:
                new = " ".join(obj['head'].split('_'))
                obj['sentence'] = obj['sentence'].replace(obj['head'], new)
                obj['head'] = new


            if "_" in obj['tail']:
                new = " ".join(obj['tail'].split('_'))
                obj['sentence'] = obj['sentence'].replace(obj['tail'], new)
                obj['tail'] = new

            sent = obj['sentence'].split(" ")

            hhead = obj['head'].split(" ")
#             print(sent, thead, hhead)
            hs = sent.index(hhead[0])
            he = []

            for i in range(len(hhead)):
                he.append(hs+i)

            thead = obj['tail'].split(" ")

            ts = sent.index(thead[0])
            te = []
            for i in range(len(thead)):
                te.append(ts+i)


            final_list[jsons[file]].append({
                "tokens": sent,
                "h": [
                    obj['head'],
                    "",
                    [he]
                ],
                "t": [
                    obj['tail'],
                    "",
                    [te]
                ]
            })
        except Exception as e:
            print("Cannot process")
            print(e)

```

```python
for key in final_list:
    print(key, len(final_list[key]))
```

```python
with open("few_rel_covid.json", "w") as outfile:
    json.dump(final_list, outfile)
```
